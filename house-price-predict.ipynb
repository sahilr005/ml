```python
df=  data[["GarageYrBlt","YearRemodAdd","YearBuilt","TotRmsAbvGrd","FullBath","TotalBsmtSF","1stFlrSF","GarageArea","GarageCars","GrLivArea","OverallQual","KitchenAbvGr","KitchenAbvGr","SalePrice","2ndFlrSF","WoodDeckSF","LotFrontage","BsmtFinSF1","Fireplaces","MasVnrArea","EnclosedPorch","KitchenAbvGr"]]
df.isnull().sum()
```

```python
df["LotFrontage"].median()
df["LotFrontage"] = df["LotFrontage"].fillna(df["LotFrontage"].median())
df["GarageArea"].fillna(df["GarageArea"].mean(),inplace=True)
df.isnull().sum()
```

```python
df[df["MasVnrArea"].isnull()]
df["MasVnrArea"] = df["MasVnrArea"].fillna(0)
df.isnull().sum()
```

```python
x= df.drop("SalePrice",axis=1)
y = df["SalePrice"]
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state= 2)
```

```python
print(x.shape,y.shape, x_train.shape , x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
print(lr.score(x_train,y_train))
print(lr.score(x_test,y_test))
# lr.predict([[2003,2003,8,2,856,856,548,2,1710,7,1,1]])
```

```python
from sklearn.linear_model import LogisticRegression
lo = LogisticRegression()
lo.fit(x_train,y_train)
print(lo.score(x_train,y_train))
print(lo.score(x_test,y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
rd = RandomForestRegressor(n_estimators=200, max_depth=60, random_state=0)
rd.fit(x_train,y_train)
print(rd.score(x_train,y_train))
print(rd.score(x_test,y_test))
```

```python
from sklearn import tree
t = tree.DecisionTreeClassifier()
t.fit(x_train,y_train)
print(t.score(x_train,y_train))
print(t.score(x_test,y_test))
```

```python
from sklearn.neural_network import MLPClassifier
nu = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1)
nu.fit(x_train,y_train)
print(nu.score(x_train,y_train))
print(nu.score(x_test,y_test))
```

```python
from sklearn.linear_model import SGDClassifier
sd = SGDClassifier( max_iter=5)
sd.fit(x_train,y_train)
print(sd.score(x_train,y_train))
print(sd.score(x_test,y_test))
```
