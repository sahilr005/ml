```python
data= pd.read_csv("/kaggle/input/titanic/train.csv")
data.head(3)
```

```python
data["Survived"].value_counts()
```

```python
data.isnull().sum()
```

```python
data["Sex"]=data["Sex"].map({"male":1,"female":2}).astype(int)
```

```python
data.Age.median()
```

```python
data["Age"].fillna(data["Age"].median(),inplace=True)
```

```python
data.shape
data.drop("Cabin",axis=1)
```

```python
data["Embarked"].value_counts()
```

```python
data["Embarked"].replace({"S":1,"C":2,"Q":3},inplace=True)
```

```python
data.info()
```

```python
data["Cabin"].value_counts()
```

```python
data.drop("Cabin",axis=1,inplace=True)
data.drop("Ticket",axis=1,inplace=True)
data.head(2)
```

```python
data.corr()["Survived"].sort_values(ascending=False)
```

```python
data.drop(["Name","PassengerId"],axis=True,inplace=True)
data.dropna(axis=0, subset=['Embarked'],inplace=True)
data.isnull().sum()
```

```python
data.info()
```

```python
x =data.drop("Survived",axis=1)
y = data["Survived"]
print(x.shape, y.shape )
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.3,random_state= 2)
```

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred[:5]
```

```python
y_test.head()
```

```python
print(lr.score(x_test,y_test))
print(lr.score(x_train,y_train))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
